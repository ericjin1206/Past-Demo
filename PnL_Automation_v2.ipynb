```python
# Define the columns to sum up
columns_to_sum = ['june_2023', 'july_2023', 'august_2023', 'september_2023',
                  'october_2023', 'november_2023', 'december_2023',
                  'january_2024', 'february_2024', 'march_2024', 'april_2024',
                  'may_2024', 'june_2024', 'july_2024']

column_sums = sales_df_sorted[columns_to_sum].fillna(0).sum()

print(column_sums)
```

```python
# Calculate the percentage for sales, direct costs, other income, and expense df
sales_df_sorted[columns_to_sum] = sales_df_sorted[columns_to_sum].div(column_sums, axis=1)
direct_costs_df_sorted[columns_to_sum] = direct_costs_df_sorted[columns_to_sum].div(column_sums, axis=1)
other_income_df_sorted[columns_to_sum] = other_income_df_sorted[columns_to_sum].div(column_sums, axis=1)
expense_df_sorted[columns_to_sum] = expense_df_sorted[columns_to_sum].div(column_sums, axis=1)
```

```python
# Function to identify months, average them, and store the output in separate datasets
def process_df(df, columns_month):
    new_df = pd.DataFrame(columns=columns_month)
    
    # Non_month columns
    new_df[['account_type', 'account_code', 'account']] = df[['account_type', 'account_code', 'account']]
    
    # Identify same months and average them
    for month in ['january', 'february', 'march', 'april', 'may', 'june', 
                  'july', 'august', 'september', 'october', 'november', 'december']:
        month_columns = [col for col in df.columns if col.startswith(month)]
        
        if month_columns:
            new_df[month] = df[month_columns].mean(axis=1)
        else:
            new_df[month] = float('nan')
    
    return new_df

# Column names for the new DataFrames
columns_month = ['account_type', 'account_code', 'account', 'january', 'february', 'march', 'april', 'may', 'june', 
                 'july', 'august', 'september', 'october', 'november', 'december']

new_df_sales = process_df(sales_df_sorted, columns_month)
new_df_direct_costs = process_df(direct_costs_df_sorted, columns_month)
new_df_other_income = process_df(other_income_df_sorted, columns_month)
new_df_expense = process_df(expense_df_sorted, columns_month)
```

```python
# Function to multiply the percentages by the corresponding revenue forecast
def multiply_forecast_with_percentage(new_df, df_revenue_forecast):
    
    # Normalize month names
    df_revenue_forecast['Month'] = df_revenue_forecast['Month'].str.strip().str.lower()
    new_df.columns = new_df.columns.str.strip().str.lower()

    result_df = pd.DataFrame(columns=['account_type', 'account_code', 'account'] + list(df_revenue_forecast['Month']))
    result_df[['account_type', 'account_code', 'account']] = new_df[['account_type', 'account_code', 'account']]
    
    for i, row in result_df.iterrows():
        # Iterate over each month_year column in result_df starting from the 4th col
        for col in result_df.columns[3:]:
            # Extract the month part 
            month_part = col.split('_')[0]
            
            if month_part in new_df.columns:
                # Find the corresponding forecast value from df_revenue_forecast
                forecast_value = df_revenue_forecast.loc[df_revenue_forecast['Month'] == col, 'Revenue Forecast'].values[0]

                result_df.at[i, col] = new_df.at[i, month_part] * forecast_value
            else:
                # Fill with NaN if no corresponding value
                result_df.at[i, col] = float('nan')

    return result_df

new_sales_forecast = multiply_forecast_with_percentage(new_df_sales, monthly_forecast)
new_direct_costs_forecast = multiply_forecast_with_percentage(new_df_direct_costs, monthly_forecast)
new_other_income_forecast = multiply_forecast_with_percentage(new_df_other_income, monthly_forecast)
new_expense_forecast = multiply_forecast_with_percentage(new_df_expense, monthly_forecast)
```

```python
def prepare_forecast_rows(forecast_df):
    
    rows_to_add = []
    
    for i, row in forecast_df.iterrows():
        for date_col in forecast_df.columns[3:]:
            
            # Change the date to standardized format
            date_obj = datetime.strptime(date_col, '%B_%Y')
            formatted_date = date_obj.strftime('%m/01/%Y')
            
            # Row dictionary
            new_row = {
                'Date': formatted_date,  
                'Account Code': row['account_code'],  
                'Account Type': row['account_type'], 
                'Account': row['account'],  
                'Amount': row[date_col],  
                # Other columns in gl_df will be left empty, status will be set to 'Budget'
                'Source': '', 'Description': '', 'Invoice Number': '', 'Reference': '',
                'Journal ID': '', 'Debit': '', 'Credit': '', 'Gross': '', 'GST': '', 'Net': '',
                'GST Rate': '', 'GST Rate Name': '', 'Cost Centre': '', 'Unique ID': '', 
                'Status': 'CP Forecast'
            }
            
            rows_to_add.append(new_row)
    
    return pd.DataFrame(rows_to_add)

new_sales_rows = prepare_forecast_rows(new_sales_forecast)
new_direct_costs_rows = prepare_forecast_rows(new_direct_costs_forecast)
new_other_income_rows = prepare_forecast_rows(new_other_income_forecast)
new_expense_rows = prepare_forecast_rows(new_expense_forecast)

# Concatenate all new rows to the existing gl_df
gl_df = pd.concat([gl_df, new_sales_rows, new_direct_costs_rows, new_other_income_rows, new_expense_rows], ignore_index=True)
```
